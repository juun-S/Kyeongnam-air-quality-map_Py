```python
import pandas as pd
```

```python
#데이터 프레임을 만들기 위한 데이터 변수 설정
stationName=[]
pm10Grade=[]
pm10Value=[]
pm25Grade=[]
pm25Value=[]
dataTime=[]
```

```python
#변수에 넣을 리스트요소넣기
for datahere in items:
    stationName.append(datahere['stationName'])
    pm10Grade.append(datahere['pm10Grade'])
    pm10Value.append(datahere['pm10Value'])
    pm25Grade.append(datahere['pm25Grade'])
    pm25Value.append(datahere['pm25Value'])
    dataTime.append(datahere['dataTime'])
```

```python
#여러 세로의 데이터프레임들을 옆으로 붙여서 하나의 데이터프레임을 만들어준다. 
datasetting = pd.concat([pd.DataFrame(stationName), 
                  pd.DataFrame(pm10Grade),
                  pd.DataFrame(pm10Value),
                  pd.DataFrame(pm25Grade),
                  pd.DataFrame(pm25Value),
                  pd.DataFrame(dataTime)],
                 axis=1)
```

```python
datasetting
```

```python
datasetting.columns=['측정소','pm10등급','pm10','pm25등급','pm25','측정시간'] 
#열마다 이름을 지정해줍니다.
```

```python
datasetting #데이터프레임완성
```

```python
import folium as fl
```

```python
station_data=pd.read_excel('경남미세먼지측정소별측정데이터_위치값포함.xlsx')
kymap=fl.Map(location=[35.2258303611, 128.50325827], tiles='Stamen Toner', zoomstart=13)
for name, lat, lng, pm10 in zip(station_data.측정소, station_data.x좌표, station_data.y좌표, station_data.pm10):
    if pm10 <= 30: #151이상 매우나쁨, 150이하 나쁨, 80이하 보통,30이하 좋음
        station_color='green'
    elif pm10 <= 80:
        station_color='yellow'
    elif pm10 <= 150 :
        station_color='orange'
    elif pm10 >= 151 :
        station_color='red'
    else:
        station_color='green'
    fl.CircleMarker([lat, lng], radius=20, color=station_color, fill=True, fill_color=station_color, fill_opacity=0.7, popup=name, tooltip=pm10).add_to(kymap)
```

```python
#초미세먼지농도에 따른 작은 원 표시
for name1, lat1, lng1, pm25 in zip(station_data.측정소, station_data.x좌표, station_data.y좌표, station_data.pm25):
    if pm25 ==None:
        station_color1='green' #15이하 좋음 , 35이하 보통, 75이하 나쁨, 76이상 매우나쁨
    elif pm25 <= 15 : 
        station_color1='green'
    elif pm25 <= 35 :
        station_color1='yellow'
    elif pm25 <= 75 : 
        station_color1='orange'
    elif pm25 >= 76 :
        station_color1='red'
    fl.CircleMarker([lat1, lng1], radius=10, color=station_color1, fill=True, fill_color=station_color1, fill_opacity=0.7, tooltip=pm25).add_to(kymap)
```
